```python
def preprocess_image(image_path, target_size=(300, 300)):
    try:
        # Load the image
        image = Image.open(image_path)
        
        # Resize the image
        image = image.resize(target_size)
        
        # Normalize pixel values
        image_array = np.array(image) / 255.0
        
        # Add batch dimension
        batched_image = np.expand_dims(image_array, axis=0)
        
        return batched_image
    except Exception as e:
        print("Error loading image:", e)
        return None
```

```python
import matplotlib.pyplot as plt
from PIL import Image

def display_detections(image_path, boxes, classes, scores, threshold=0.5):
    # Load the original image for visualization
    image = Image.open(image_path)
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    ax = plt.gca()

    for i, box in enumerate(boxes):
        if scores[i] >= threshold:
            y_min, x_min, y_max, x_max = box
            rect = plt.Rectangle((x_min * image.width, y_min * image.height),
                                 (x_max - x_min) * image.width,
                                 (y_max - y_min) * image.height,
                                 linewidth=2, edgecolor='red', facecolor='none')
            ax.add_patch(rect)
            plt.text(x_min * image.width, y_min * image.height,
                     f"Class: {int(classes[i])}, Score: {scores[i]:.2f}",
                     color='white', backgroundcolor='red')
    plt.axis('off')
    plt.show()

sample_image_path = r"C:\Users\Hp\Desktop\busy-street-scene-midtown-manhattan-nyc-ny-usa-AYGEY1.jpg"
display_detections(sample_image_path, bounding_boxes[0], class_labels[0], confidence_scores[0])
```
